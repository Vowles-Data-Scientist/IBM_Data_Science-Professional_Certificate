### This is a basic calculator program using Python 3.9

<br>

```python
a = int(input('please enter the first number: '))
b = int(input('please enter the second number: '))
print('thank you')
oper = input('please enter one operation (+, -, *, /, //, %, **): ')
if (oper == '+'):
    print(a + b)
elif (oper == '-'):
    print(a - b)
elif (oper == '-'):
    print(a - b)
elif (oper == '*'):
    print(a * b)
elif (oper == '/'):
    print(a / b)
elif (oper == '//'):
    print(a // b)
elif (oper == '%'):
    print(a % b)
elif (oper == '**'):
    print(a ** b)
print('thank you, here is your answer')
```
